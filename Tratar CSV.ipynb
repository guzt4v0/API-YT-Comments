```python
import pandas as pd
```

```python
df = pd.read_csv('youtube-comments.csv') #<---- CSV gerado 
```

```python
df.head()
```

```python
csvList = df['Comment'].tolist()
```

```python
type(csvList)
```

```python
from collections import Counter
```

```python
count = dict(Counter(word for sentence in csvList for word in sentence.split()))
```

```python
count
```

```python
type(count)
```

```python
def sort_dict_by_value(count, reverse = False):
    return dict(sorted(count.items(), key = lambda x: x[1], reverse = reverse))
```

```python
dic_order = sort_dict_by_value(count, True)
```

```python
dic_order
```

```python
entities_to_remove = ('que', 'de', 'e', 'a', '√©', 'o', 'se', 'com', 'do', 'os', 'em', 'por', 'as', 'da', 'mas', 'pra', 'na', 'no', 'para')
for x in entities_to_remove:
    if x in dic_order:
        del dic_order[x]
```

```python
dic_order
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud

wordcloud = WordCloud(width = 1000, height = 500).generate_from_frequencies(dic_order)

plt.figure(figsize=(15,8))
plt.imshow(wordcloud);
```
